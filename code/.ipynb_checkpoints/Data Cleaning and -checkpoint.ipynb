#### Test For Stationarity - Augmented Dickey-Fuller Test

```python
def adf_test(series,title=''):
    result = adfuller(series.dropna(),autolag='AIC')
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)
    
    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          
    
    if result[1] <= 0.05:
        print("Data is stationary")
    else:
        print("Data is non-stationary")
```

```python
# Test on original dataset, p value above 0.05 showing the dataset reject H1 and is non stationary
adf_test(final_df['eur/usd'])
```

```python
final_df['euro_1st_diff'] = final_df['eur/usd'].diff(1)
final_df['euro_1st_diff'] = final_df['euro_1st_diff'].bfillna
adf_test(final_df['euro_1st_diff'])
```

```python
# Create a 80:20 train/test split data
# Used month of Nov 2019 as train data, and 1st week of Dec as test data
train = final_df.iloc[0:892]
test = final_df.iloc[892:]
```

```python
# Instantiate an AR(1) model on training data.
# Note this training data consists of the 
# time periods 0 through 40.
ar1 = ARIMA(endog = train['euro_1st_diff'],
            order = (1, 0, 0)) # (p, d, q)

# Fit AR(1) model.
model = ar1.fit()

# Generate predictions based on test set.
# Start at time period 41 and end at 49.
preds = model.predict(start=892, end=1113)

# Evaluate predictions.
mean_squared_error(test['euro_1st_diff'], preds)
```

```python
# Plot data.
plt.figure(figsize=(16,6))

# Plot training data in blue.
plt.plot(train['eur/usd'], color = 'blue')

# Plot testing data in orange.
plt.plot(test['eur/usd'], color = 'orange')

# Plot predicted values in green.
plt.plot(preds, color = 'green')

plt.title(label = 'EUR/USD', fontsize=18)
plt.show();
```
